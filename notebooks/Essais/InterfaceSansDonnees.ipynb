# Interface sans les données

```python
import pandas as pd
import numpy as np
import plotly.express as px
from ipywidgets import interact, widgets

jours = range(1, 101)  # 100 jours
ventes = np.random.randint(1000, 5000, size=100)  # ventes aléatoires

sales_per_day = pd.DataFrame({
    "day": jours,
    "sales": ventes
})

def afficher_ventes(nb_jours=30):
    # On prend les derniers nb_jours
    data = sales_per_day.tail(nb_jours)
    # Graphe interactif Plotly
    fig = px.line(data, x="day", y="sales",
                  title=f"Ventes des {nb_jours} derniers jours",
                  labels={"day":"Jour", "sales":"Ventes"})
    fig.show()


interact(afficher_ventes,
         nb_jours=widgets.IntSlider(min=10, max=len(sales_per_day), step=10, value=30))
```
