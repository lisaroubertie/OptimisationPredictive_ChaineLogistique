## 2. Ventes par magasin au fil du temps
On visualise l'évolution des ventes pour chacun des 10 magasins avec un lissage sur 7 jours pour filtrer le bruit quotidien et faire ressortir les tendances de fond.

```python
import matplotlib.dates as mdates

sales_by_store = sales.groupby('store_id')[day_cols].sum()
stores = sales_by_store.index.tolist()
dates  = [day_to_date[d] for d in day_cols if d in day_to_date]

fig, axes = plt.subplots(len(stores), 1, figsize=(18, len(stores) * 3), sharex=True)

for i, store in enumerate(stores):
    ax       = axes[i]
    y_raw    = sales_by_store.loc[store].values[:len(dates)]
    y_smooth = pd.Series(y_raw).rolling(window=7, center=True).mean()
    
    ax.fill_between(dates, y_raw, alpha=0.2, color='steelblue')
    ax.plot(dates, y_raw, color='steelblue', alpha=0.3, linewidth=0.5)
    ax.plot(dates, y_smooth, color='steelblue', linewidth=2, label='Moyenne 7j')
    ax.set_title(f'Magasin {store}', fontsize=12, fontweight='bold')
    ax.set_ylabel('Ventes totales')
    ax.legend(fontsize=8)
    ax.grid(True, alpha=0.3)
    ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m'))
    ax.xaxis.set_major_locator(mdates.MonthLocator(interval=3))

plt.suptitle('Ventes par magasin au fil du temps', fontsize=16, fontweight='bold', y=1.01)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

Ces graphes révèlent des comportements très hétérogènes entre magasins.

**Californie** : CA_3 domine largement avec des volumes entre 4000 et 8000 ventes par jour, tandis que CA_4 est le moins actif avec seulement 1500 à 2500 ventes. CA_1 montre la progression la plus régulière et stable. CA_2 reste stable jusqu'en 2015 puis connaît une hausse soudaine inexpliquée par les données seules.

**Texas** : TX_2 est le plus performant avec des volumes entre 3000 et 5000. TX_1 et TX_3 sont similaires et stables autour de 2500-3500.

**Wisconsin** : Le fait le plus marquant est le basculement brutal de WI_1 et WI_2 vers mi-2012, où les ventes doublent quasi instantanément: WI_1 passe d'environ 1500 à 3000 et WI_2 d'environ 2000 à 4000. En revanche WI_3 reste stable sur toute la période malgré le fait qu'il soit dans le même état, ce qui montre que les disparités sont propres à chaque magasin et non à l'état.

**Conclusion** : Ces disparités confirment qu'on ne peut pas modéliser par état seulement: il faudra intégrer le store_id comme variable dans notre modèle de prévision. Ces graphes permettent de visualiser la dynamique temporelle mais comparer les volumes annuels entre magasins reste difficile. On utilise donc une heatmap pour avoir une vue synthétique en un seul coup d'œil.

## 3. Heatmap des ventes totales par magasin et par année

Les courbes précédentes nous ont permis de visualiser la dynamique temporelle détaillée de chaque magasin. Cependant comparer 10 magasins simultanément reste difficile. On utilise donc une heatmap qui permet de voir les volumes annuels de tous les magasins en un seul coup d'œil grâce au code couleur, et d'avoir les chiffres exacts. 
Note : les données 2016 ne couvrent que janvier à juin, elles sont donc moins élevées que les autres années mais restent incluses pour montrer la tendance en cours.

```python
# On récupère l'année pour chaque jour
calendar['year'] = calendar['date'].dt.year
year_day = calendar[['d', 'year']].set_index('d')


heatmap_data = {}
for store in sorted(sales['store_id'].unique()):
    store_sales = sales[sales['store_id'] == store][day_cols].sum()
    yearly = {}
    for year in [2011, 2012, 2013, 2014, 2015, 2016]:
        days = [d for d in year_day[year_day['year'] == year].index if d in store_sales.index]
        yearly[year] = store_sales[days].sum()
    heatmap_data[store] = yearly

df_heatmap = pd.DataFrame(heatmap_data).T

plt.figure(figsize=(10, 7))
sns.heatmap(df_heatmap, annot=True, fmt='.0f', cmap='YlOrRd')
plt.title('Ventes totales par magasin et par année', fontsize=13, fontweight='bold')
plt.xlabel('Année')
plt.ylabel('Magasin')
plt.tight_layout()
plt.show()
```

**Interprétation :**

La heatmap met en évidence des écarts significatifs de performance entre les magasins. CA_3 se démarque clairement comme le magasin le plus performant avec des volumes dépassant 2,3 millions de ventes en 2014, soit presque deux fois plus que CA_1 qui est pourtant le deuxième magasin californien. À l'inverse, CA_4, CA_2 et TX_1 affichent des volumes largement inférieurs sur toute la période. WI_1 reste le magasin le moins actif du Wisconsin.

On observe une tendance à la hausse entre 2011 et 2015 pour la quasi-totalité des magasins. Cette progression est particulièrement marquée pour WI_2 dont les volumes ont presque triplé sur la période, contrairement à WI_1 pourtant dans le même état. Cela suggère un potentiel développement commercial significatif mais les données seules ne permettent pas de le confirmer.

Ces disparités confirment qu'un modèle qui ignorerait les différences entre magasins serait trop généraliste. Il faudra intégrer store_id et state_id comme variables d'entrée du modèle.

## 5. Comparaison des ventes entre états

Les disparités observées entre magasins soulèvent une question : ces écarts sont-ils propres à chaque magasin ou reflètent-ils des disparités plus larges à l'échelle des états ? On sépare ici les ventes par état sur l'ensemble de la période.

```python
# Ventes totales par état et par jour
sales_by_state = sales.groupby('state_id')[day_cols].sum()
dates = [day_to_date[d] for d in day_cols if d in day_to_date]

colors_state = {'CA': '#2196F3', 'TX': '#F44336', 'WI': '#4CAF50'}

fig, axes = plt.subplots(2, 1, figsize=(18, 10))

# Graphe 1 : séries temporelles lissées sur 30 jours
for state in sales_by_state.index:
    y = sales_by_state.loc[state].values[:len(dates)]
    y_smooth = pd.Series(y).rolling(30, center=True).mean()
    axes[0].plot(dates, y_smooth, linewidth=2.5, 
                 label=f'Etat {state}', color=colors_state.get(state))

axes[0].set_title('Ventes totales par état (moyenne lissée 30 jours)', fontsize=13, fontweight='bold')
axes[0].set_ylabel('Ventes totales')
axes[0].legend(fontsize=11)
axes[0].grid(True, alpha=0.3)

# Graphe 2 : ventes annuelles par état
calendar['year'] = calendar['date'].dt.year
year_day = calendar[['d', 'year']].set_index('d')

state_year = {}
for state in sales_by_state.index:
    row = sales_by_state.loc[state]
    yearly = {}
    for year in [2011, 2012, 2013, 2014, 2015, 2016]:
        days = [d for d in year_day[year_day['year'] == year].index if d in row.index]
        yearly[year] = row[days].sum()
    state_year[state] = yearly

df_year = pd.DataFrame(state_year).T
df_year.T.plot(kind='bar', ax=axes[1], 
               color=[colors_state.get(s) for s in df_year.index], width=0.7)

axes[1].set_title('Ventes annuelles par état', fontsize=13, fontweight='bold')
axes[1].set_ylabel('Ventes totales')
axes[1].set_xlabel('Année')
axes[1].legend(title='Etat', fontsize=10)
plt.setp(axes[1].xaxis.get_majorticklabels(), rotation=0)

plt.suptitle('Comparaison des ventes entre états', fontsize=16, fontweight='bold')
plt.tight_layout()
plt.show()
```

```python
moyenne_par_etat = sales_by_state.mean(axis=1)
print(moyenne_par_etat)
```

**Interprétation :**

La Californie domine sur toute la période avec en moyenne 15 042 ventes par jour contre 9 906 pour le Texas et 9 532 pour le Wisconsin. Cette domination confirme que les écarts observés entre CA_3 et les autres magasins ne sont pas propres à ces magasins, ils reflètent une dynamique géographique plus profonde.

Malgré le fait que le Texas et le Wisconsin montrent des volumes similaires, ils restent bien en dessous de la Californie. ## 6. Corrélation entre magasins

Les analyses précédentes ont mis en évidence des disparités importantes entre magasins en termes de volumes et de dynamique temporelle. On cherche maintenant à mesurer à quel point les magasins évoluent de façon similaire dans le temps.

Une forte corrélation suggère des facteurs communs comme la saisonnalité nationale. Une faible corrélation indique des dynamiques locales propres à chaque magasin.

On utilise deux niveaux de lecture complémentaires : une heatmap pour comparer toutes les paires en un seul coup d'œil, puis des scatter plots pour visualiser concrètement la qualité de la corrélation.

```python
sales_by_store_daily = sales.groupby('store_id')[day_cols].sum().T
sales_by_store_daily.index = [day_to_date.get(d) for d in sales_by_store_daily.index]
sales_by_store_daily = sales_by_store_daily.dropna()

# Matrice de corrélation 
corr_matrix = sales_by_store_daily.corr()

plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, fmt='.3f', cmap='RdYlGn',
            vmin=0.8, vmax=1, linewidths=0.5, square=True,
            cbar_kws={'label': 'Corrélation de Pearson'})
plt.title('Corrélation entre magasins', fontsize=13, fontweight='bold')
plt.tight_layout()
plt.show()
```

**Interprétation :**

Tous les magasins sont positivement corrélés entre eux avec des valeurs allant de 0.311 à 0.906. Les corrélations les plus fortes s'observent généralement au sein d'un même état : CA_1 et CA_3 affichent 0.906, TX_1 et TX_3 atteignent 0.878.

Cependant cette tendance n'est pas systématique. WI_1 et WI_3 n'affichent que 0.311, ce qui indique que l'appartenance à un même état ne suffit pas à expliquer les similarités de comportement entre magasins.

Ces résultats confirment que les magasins ne sont ni totalement indépendants ni parfaitement synchronisés. Le modèle devra donc tenir compte à la fois des tendances globales et des spécificités de chaque magasin.

### Scatter plots intra-état
On compare les magasins au sein d'un même état pour voir si l'appartenance géographique suffit à expliquer les similarités.

```python
from scipy import stats

paires_intra = [('CA_1', 'CA_3'), ('TX_1', 'TX_3'), ('WI_1', 'WI_3')]

fig, axes = plt.subplots(1, 3, figsize=(18, 5))

for i, (s1, s2) in enumerate(paires_intra):
    axes[i].scatter(sales_by_store_daily[s1], sales_by_store_daily[s2],
                    alpha=0.3, color='steelblue', s=10)
    slope, intercept, r, p, _ = stats.linregress(sales_by_store_daily[s1],
                                                  sales_by_store_daily[s2])
    x_line = np.linspace(sales_by_store_daily[s1].min(),
                         sales_by_store_daily[s1].max(), 100)
    axes[i].plot(x_line, slope * x_line + intercept, color='red',
                 linewidth=2, label=f'R² = {r**2:.3f}')
    axes[i].set_xlabel(f'Ventes {s1}', fontsize=10)
    axes[i].set_ylabel(f'Ventes {s2}', fontsize=10)
    axes[i].set_title(f'{s1} vs {s2}', fontsize=11, fontweight='bold')
    axes[i].legend(fontsize=10)

plt.suptitle('Corrélation intra-état', fontsize=13, fontweight='bold')
plt.tight_layout()
plt.show()
```

### Scatter plots inter-états
On compare les représentants de chaque état pour voir si les corrélations dépassent les frontières géographiques.

```python
paires_inter = [('CA_3', 'TX_2'), ('CA_3', 'WI_2'), ('TX_2', 'WI_2')]

fig, axes = plt.subplots(1, 3, figsize=(18, 5))

for i, (s1, s2) in enumerate(paires_inter):
    axes[i].scatter(sales_by_store_daily[s1], sales_by_store_daily[s2],
                    alpha=0.3, color='steelblue', s=10)
    slope, intercept, r, p, _ = stats.linregress(sales_by_store_daily[s1],
                                                  sales_by_store_daily[s2])
    x_line = np.linspace(sales_by_store_daily[s1].min(),
                         sales_by_store_daily[s1].max(), 100)
    axes[i].plot(x_line, slope * x_line + intercept, color='red',
                 linewidth=2, label=f'R² = {r**2:.3f}')
    axes[i].set_xlabel(f'Ventes {s1}', fontsize=10)
    axes[i].set_ylabel(f'Ventes {s2}', fontsize=10)
    axes[i].set_title(f'{s1} vs {s2}', fontsize=11, fontweight='bold')
    axes[i].legend(fontsize=10)

plt.suptitle('Corrélation inter-états', fontsize=13, fontweight='bold')
plt.tight_layout()
plt.show()
```

Les scatter plots révèlent des comportements très contrastés. En intra-état, CA_1 vs CA_3 et TX_1 vs TX_3 montrent des points bien groupés autour de la droite avec quelques outliers ponctuels. Le cas le plus surprenant est WI_1 vs WI_3 avec R^2=0.097: la droite est presque plate et les points totalement éparpillés, ce qui confirme que deux magasins du même état peuvent évoluer de façon complètement indépendante.

En inter-états, CA_3 vs TX_2 reste la corrélation la plus forte avec R^2=0.615 malgré des états différents, plus forte que WI_1 vs WI_3 qui sont pourtant dans le même état. CA_3 vs WI_2 montre deux nuages distincts, probablement liés à la transition de WI_2 vers 2013. TX_2 vs WI_2 avec R^2=0.243 confirme une relation très faible entre ces deux magasins.

Ces observations renforcent la conclusion que la modélisation devra se faire au niveau du magasin et non de l'état.

Note : les représentants ont été choisis selon les volumes les plus significatifs par état. D'autres paires auraient pu donner des résultats différents: WI_1 vs WI_2 par exemple affiche 0.790, bien supérieur à WI_1 vs WI_3.


## Piste d'amélioration suite au mi-projet - Impact du prix sur les ventes par magasin

Suite au rapport de mi-projet, nous avons identifié que l'analyse du prix pourrait enrichir notre compréhension des variations de ventes par magasin. On analyse ici si les variations de prix influencent les volumes vendus sur les produits les plus vendus par magasin.

```python
prix_moyen_etat = sell_prices.groupby('store_id')['sell_price'].mean().reset_index()
prix_moyen_etat['state'] = prix_moyen_etat['store_id'].str[:2]
prix_moyen_state = prix_moyen_etat.groupby('state')['sell_price'].mean()

plt.figure(figsize=(8, 5))
plt.bar(prix_moyen_state.index, prix_moyen_state.values, color=['#2196F3', '#F44336', '#4CAF50'])
plt.title('Prix moyen par état', fontsize=13, fontweight='bold')
plt.ylabel('Prix moyen ($)')
plt.xlabel('État')
plt.tight_layout()
plt.show()

print(prix_moyen_state)
```

```python
prix_moyen_magasin = sell_prices.groupby('store_id')['sell_price'].mean().sort_values()

plt.figure(figsize=(10, 5))
plt.bar(prix_moyen_magasin.index, prix_moyen_magasin.values, color='steelblue')
plt.title('Prix moyen par magasin', fontsize=13, fontweight='bold')
plt.ylabel('Prix moyen (dollard usa)')
plt.xlabel('Magasin')
plt.tight_layout()
plt.show()

print(prix_moyen_magasin)
```

Cette analyse montre que les prix moyens sont quasi identiques entre tous les magasins et tous les états, avec des valeurs comprises entre 4.37 et 4.46 dollards. Walmart applique donc une politique de prix uniforme. 

Cependant cette analyse reste limitée car elle agrège tous les produits sans distinction de catégorie. Une analyse plus fine nécessiterait de comparer les prix produit par produit ce qui dépasse le scope de cette section et relève davantage de l'analyse par produit.
